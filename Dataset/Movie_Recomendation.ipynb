## Data Transformation

```python
Movies_DS.iloc[0].genres
```

```python
import ast
```

```python
def ConvertToList(object):
    Li = [] 
    for i in ast.literal_eval(object):
        Li.append(i['name']) 
    return Li
```

```python
# ConvertToList(Movies_DS.iloc[0].genres)
```

```python
Movies_DS['genres'] = Movies_DS['genres'].apply(ConvertToList)
```

```python
Movies_DS.head()
```

```python
Movies_DS['keywords'] = Movies_DS['keywords'].apply(ConvertToList)
```

```python
Movies_DS.head()
```

```python
Movies_DS.iloc[0].cast
```

```python

def ConvertCastList(object):
    Li = [] 
    counter = 0
    for i in ast.literal_eval(object):
        if counter == 3:
            break
        Li.append(i['name']) 
    return Li 
```

```python
Movies_DS['cast'] = Movies_DS['cast'].apply(ConvertCastList)
```

```python
Movies_DS.head()
```

```python
# Movies_DS['crew'][0]
```

```python

def Fetch_Director_Name(object):
    Li = [] 
    for i in ast.literal_eval(object):
        if i['job'] == 'Director':
            Li.append(i['name'])
            break
    return Li
```

```python
Movies_DS['crew'] = Movies_DS['crew'].apply(Fetch_Director_Name)
```

## Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python

def Stem(text):
    Li = [] 
    for i in text.split():
        Li.append(ps.stem(i))
    return " ".join(Li)
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

## Forming Similarity Matrix

* Find the Distance/Similarity for each vector with respect to all other Vector

```python
# from sklearn.metrics.pairwise import cosine_distances 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

## Finding Recommendations

```python
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
def Recommend_Movie(current_movie):
    movie_idx = new_df[new_df['title'] == current_movie].index[0]
    distances = similarity[movie_idx] 
    rec_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6] 

    for i in rec_list:
        # print(i[0]) 
        
        print(new_df.iloc[i[0]].title)
    
    
```
